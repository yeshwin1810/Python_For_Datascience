```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
data = pd.read_csv(r'C:\Users\tkaks\Downloads\archive\climate_change_impact_on_agriculture_2024.csv')

plt.plot(data['Crop_Yield_MT_per_HA'])
plt.plot(data['Average_Temperature_C'])
plt.title("Scatter plot")
plt.ylabel('Country')
plt.show()
```

```python
plt.bar(data['Country'], data['Crop_Yield_MT_per_HA'],color = 'green')
plt.title("Bar Chart")
plt.ylabel('Crop Type',fontsize=12)
plt.xlabel('Country',fontsize=12)
plt.show()
```

```python
plt.hist(data['Total_Precipitation_mm'],color = 'red',edgecolor = 'black') 
plt.title("Histogram")
plt.show()
```
