```python
# importing the modules
from bokeh.plotting import figure, output_file, show
from bokeh.palettes import magma
import pandas as pd


# instantiating the figure object
graph = figure(title = "Bokeh Scatter Graph")

# reading the database
data = pd.read_csv(r"C:\Users\keyes\Downloads\Customer Purchasing Behaviors.csv")

color = magma(238)

# plotting the graph
graph.scatter(data['region'], data['purchase_frequency'], color=color)

# displaying the model
show(graph)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# reading the database
data = pd.read_csv(r"C:\Users\keyes\Downloads\Customer Purchasing Behaviors.csv")

# Scatter plot with day against tip
plt.scatter(data['age'], data['region'], c=data['loyalty_score'], s=data['purchase_amount'])

# Adding Title to the Plot
plt.title("Scatter Plot")

# Setting the X and Y labels
plt.xlabel('age')
plt.ylabel('frequency')

plt.colorbar()

plt.show()
```
