```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv(r"C:\Users\tkaks\Downloads\Toyota.csv",index_col=0,na_values=["??","????"])
data.head()
```

```python
data.dropna(axis = 0,inplace = True)
print(data.shape)
```

```python
##Scatter plot
```

```python
plt.scatter(data['Age'],data['Price'],color='red')
plt.title('Scatter plot of price vs age of the car')
plt.xlabel('Age(month)')
plt.ylabel('Price(Euros)')
plt.show()
```

```python
plt.hist(data['KM'],color = 'green',edgecolor = 'white')
plt.title('Histogram of Kilometer')
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
```

```python
col = data['FuelType'].value_counts()
fuelType = ('Petrol','Disel','CNG')
index = np.arange(len(fuelType))
plt.bar(index,col,color = ['red','blue','green'])
plt.title('Bar graph')
plt.xlabel('FuelType')
plt.ylabel('Frequency')
plt.xticks(index,fuelType)
plt.show()
```

```python
sns.set(style = 'darkgrid')
sns.regplot(x  = data['Age'],y = data['Price'])
```
